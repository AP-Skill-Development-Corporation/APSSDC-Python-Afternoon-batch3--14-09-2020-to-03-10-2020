##### Data Visualization
* Matplotlib is one of the library of our python
* This is used to visualize the data in 2D format
* it can be used in python script,shell,web application server


```python
from matplotlib import pyplot
```

### Types of graphs
* line graph
* bar graph
* histogram
* scatter graph
* areaplot
* piechart


```python
#line graph
from matplotlib import pyplot as plt
plt.plot([1,2,3,4],[5,6,7,8])
plt.show()
```

```python
x=[5,2,7]
y=[2,16,4]
plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line-graph")
plt.show()
```

```python
#applying styles
x=[5,2,7]
y=[2,16,4]
a=[6,9,11]
b=[21,54,3]
plt.plot(x,y,"g",label="lineone",linewidth=4)
plt.plot(a,b,"R",label="linetwo",linewidth=5)
plt.title("Line-graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.grid(color='b')
plt.show()
```

```python
x=np.linspace(0,10)
y=np.exp(-x)
plt.plot(x,y)
plt.savefig("linegraph.svg")
plt.savefig("linegraph.jpg")
plt.show()
```

```python
#bar graph
plt.bar([10,20,30],['c','C++','python'])
plt.title("bar graph")
plt.xlabel("values")
plt.ylabel("languages")
plt.show()
```

```python
#applying styles
plt.bar([10,30,50],[20,40,60],color="m",label="values",width=10)
plt.bar([70,90,50],[70,80,90],color="y",label="Languages",width=10)
plt.legend()
plt.title("bar-graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.savefig("bargraph.svg")
plt.savefig("bargraph.jpg")
plt.show()
```

```python
np.random.normal(5,4,10)
```

```python
from numpy.random import normal
x=normal(size=200)
plt.hist(x,bins=50,color="b")
plt.title("histogram")
plt.savefig("histogram.svg")
plt.show()
```

```python
#population ages
age=[18,22,24,26,28,30]
b=[0,10,20,30,50,60]
plt.hist(age,b,rwidth=0.5)
plt.show()
```

```python
#scattergraph
plt.scatter([2,3,4,6],[7,8,9,10])
plt.show()
```

```python
a=[1,1.5,2,2.5,3,3.5,4,4.5]
b=[10,15,20,25,30,35,40,45]
a1=[6,6.5,7,7.5,8,8.5,9,9.5]
b1=[10,20,30,40,50,17,64,12]
plt.scatter(a,b,label="high income low saving",color='r')
plt.scatter(a1,b1,label="low saving high income",color="b")
plt.legend()
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("scatter graph")
#plt.grid()
plt.savefig("scatter.svg")
plt.savefig("scatter.jpg")
plt.show()
```
